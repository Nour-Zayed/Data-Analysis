```python
df['Type'] = df['Type'].str.replace('\r\n', 'other')

df['Type'].value_counts()
```

```python
#Removing the word 'Reviews' from the column review

df['Reviews'] = df['Reviews'].str.replace('Reviews', '')
df['Reviews']
```

```python
df['Information'] = (
    df['Information']
    .str.replace(r'[\r\n\t]+', ' ', regex=True)  # Replace any combination of \r, \n, or \t
    .str.strip()
)

df.Information.value_counts()
```

```python
df
```

```python
#Categorizing the Column of Effective

effectiveness = []

for score in df['Effective']:
    if score < 1.0 : effectiveness.append('Very Uneffective')
    elif score < 2.0 : effectiveness.append('Uneffective')
    elif score < 3.0 : effectiveness.append('Partly Effective')
    elif score < 4.0 : effectiveness.append('More Than Effective')
    elif score <= 5.0 : effectiveness.append('Very Effective')

df['level_of_effectiveness'] = effectiveness
```

```python
#Categorizing the column ease of use

easeofuse = []

for score in df['EaseOfUse']:
    if score < 1.0 : easeofuse.append('Very Difficult')
    elif score < 2.0 : easeofuse.append('Difficult')
    elif score < 3.0 : easeofuse.append('Normal')
    elif score < 4.0 : easeofuse.append('Easy')
    elif score <= 5.0 : easeofuse.append('Very Easy')

df['level_of_difficulty'] = easeofuse
```

```python
satisfaction_level = []

for score in df['Satisfaction']:
    if score < 1.0 : satisfaction_level.append('Very Unsatisfied')
    elif score < 2.0 : satisfaction_level.append('Unsatisfied')
    elif score < 3.0 : satisfaction_level.append('Partly Satisfied')
    elif score < 4.0 : satisfaction_level.append('More Than Satisfied')
    elif score <= 5.0 : satisfaction_level.append('Very Satisfied')

df['level_of_satisfaction'] = satisfaction_level
```

```python
df.groupby("Condition").size().sort_values().plot(kind = "bar" , width = 1 ,color = sns.color_palette('Paired'))
plt.xlabel("Condition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Type', y='level_of_effectiveness')

plt.title('Bar Chart of Type vs Effectiveness', fontsize=16)
plt.xlabel('Type', fontsize=12)
plt.ylabel('level_of_effectiveness', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Type', hue='level_of_difficulty', multiple="stack", kde=True)

plt.title('Histogram of Reviews vs level_of_difficulty', fontsize=16)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
top_10_drugs = df.groupby('Drug')['Effective'].mean().sort_values(ascending=False).head(10)

# Create a bar chart for the top 10 drugs
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_drugs.index, y=top_10_drugs.values)

# Add title and labels
plt.title('Top 10 Drugs Based on Effectiveness', fontsize=16)
plt.xlabel('Drug', fontsize=12)
plt.ylabel('Average Effectiveness', fontsize=12)
plt.xticks(rotation=45)
# Show the plot
plt.tight_layout()
plt.show()
```

## ** Report **


This analysis aims to explore the relationship between various drugs and their effectiveness, with a focus on identifying the top 10 drugs based on average effectiveness. The dataset contains information about multiple drugs, their respective effectiveness ratings, and several other parameters such as reviews, ease of use, satisfaction, and level of difficulty. For this report, we specifically examine the Effectiveness of the drugs and provide insights regarding the top-performing drugs in terms of this metric.

 **Methodology**

Data Grouping: The data was grouped by the Drug column, and for each drug, the average effectiveness score was calculated.
Sorting: The drugs were sorted based on the average effectiveness score in descending order to identify the top-performing drugs.
Visualization: A bar chart was created to visualize the top 10 drugs in terms of their effectiveness ratings.


**Data Overview**


The dataset includes the following key columns:

Condition: The medical condition for which the drug is prescribed.
Drug: The name of the drug.
Indication: The specific indication or purpose of the drug.
Type: The classification of the drug (e.g., generic, brand, etc.).
Reviews: The number of user reviews for the drug.
Effective: The effectiveness rating of the drug (scale or score).
EaseOfUse: The ease of use rating of the drug.
Satisfaction: The user satisfaction rating for the drug.
Information: Information-related ratings, potentially related to knowledge about the drug.
Level of Effectiveness: A categorical representation of the drug's effectiveness.
Level of Difficulty: How difficult the drug is to use.
Level of Satisfaction: The satisfaction level based on users' feedback.
