#### Which company offers the best specifications at a reasonable price?

```python

import re

# Function to convert storage to GB
def convert_storage(value):
    match = re.match(r'(\d+)(TB|GB)', str(value))
    if match:
        num, unit = match.groups()
        num = int(num)
        return num * 1024 if unit == "TB" else num  #Convert TB to GB
    return None

df["Storage_GB"] = df["Storage"].apply(convert_storage)

# Ensure that columns contain only numbers.
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["RAM__GB_"] = pd.to_numeric(df["RAM__GB_"], errors="coerce")
df["Battery_Life__hours_"] = pd.to_numeric(df["Battery_Life__hours_"], errors="coerce")

# Data collection by brand
brand_specs = df.groupby("Brand").agg({
    "Price": "mean",
    "RAM__GB_": "mean",
    "Storage_GB": "mean",
    "Battery_Life__hours_": "mean"
}).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x="Brand", y="Price", data=brand_specs, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Average laptop Price per brand")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="RAM__GB_", y="Price", data=df, alpha=0.6)
sns.regplot(x="RAM__GB_", y="Price", data=df, scatter=False, color="red")
plt.title("The relationship between RAM and the price of the laptop")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error
```

```python


# Encoding categorical variables using LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding on categorical columns
df['Brand'] = label_encoder.fit_transform(df['Brand'])
df['Processor'] = label_encoder.fit_transform(df['Processor'])
df['GPU'] = label_encoder.fit_transform(df['GPU'])
df['Operating_System'] = label_encoder.fit_transform(df['Operating_System'])

# Extracting screen width from the resolution (taking the width)
# Ensure that the 'Resolution' column is treated as strings
df['Resolution'] = df['Resolution'].astype(str).apply(lambda x: int(x.split('x')[0]))

# Features (independent variables) and target (Price)
X = df[['Brand', 'Processor', 'RAM__GB_', 'Storage_GB', 'GPU', 'Screen_Size__inch_', 'Resolution', 'Battery_Life__hours_', 'Weight__kg_', 'Operating_System']]
y = df['Price']

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Linear Regression Model
model = LinearRegression()
model.fit(X_scaled, y)

coefficients = model.coef_

# Display the coefficients with their corresponding features
coeff_df = pd.DataFrame(list(zip(X.columns, coefficients)), columns=['Feature', 'Coefficient'])
print(coeff_df)

correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

The code processes device data by encoding categorical columns like Brand, Processor, GPU, and Operating_System into numerical values using LabelEncoder. It then scales the features using StandardScaler for consistency and fits a Linear Regression model to predict the Price based on various factors. Finally, the code calculates and displays the model's coefficients along with a heatmap of feature correlations to better understand the relationships between variables and the price.
